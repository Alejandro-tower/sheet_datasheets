```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
from tqdm import tqdm
```

### Instrucciones, el presente scrip grafica usando tranformaciones lineales; 
* Dado que los ejes pueden cambiar su punto de vista, se puede cambiar la variable en la cual rota, por defecto este script hace rotar alrededor de Z, sin embargo se pueden cambiar los nombres de las variables para lograr el efecto visual(se puede cambiar la etiqueta con setlabel).
* Los comandos ax.set definen los rangos de los ejes que se visualizaran.
* Los colores se cambian con el comando: cmap=plt.cm.[barra de colores], por ejemplo: cmap=plt.cm.autumm; los colores disponibles estan al final de este notebook
* Alpha es la transparencia de la figura, con intervalo [0:1], por ejemplo  alpha=0.41 y servira para poner la superficie interior con mayor solides y la exterior con mayor transparencia.


### Solido de revolución acotado por y=e^x; y=x, con x: [0:1] alrededor de x

```python
%matplotlib notebook
box=np.e # set draw area as box^3
fig = plt.figure(figsize=(7,7)) # set fig with its size 10 x 10 inch
ax = fig.add_subplot(111,projection='3d') # creates an additional axis to the standard 2D
ax.set_xlim(-box,box) # set x-range
ax.set_ylim(-box,box) # set y-range
ax.set_zlim(-box,box) # set z-range


# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(1, np.e, 50)
p = np.linspace(0.001,2*np.pi, 50)
R, P = np.meshgrid(r, p)
Z = (np.log(R))
# Express the mesh in the cartesian system.
X, Y = R*np.cos(P), R*np.sin(P)
#
r1 = np.linspace(0, 1, 50)
p1 = np.linspace(0, 2*np.pi, 50)
R1, P1 = np.meshgrid(r1, p1)
X1, Y1 = R1*(np.cos(P1)), R1*(np.sin(P))
Z1= R1


# Plot the surface.
heat=ax.plot_surface(X1, Y1, Z1, cmap=plt.cm.OrRd, alpha=1)
surf=ax.plot_surface(X, Y, Z, cmap=plt.cm.YlGn, alpha=0.55)


# Tweak the limits and add latex math labels.
#ax.set_zlim(0, 1)
ax.set_xlabel(r"Z",fontsize=20) # set x-label
ax.set_ylabel(r"y",fontsize=20) # set y-label
ax.set_zlabel(r"x",fontsize=20) # set z-label

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
fig.colorbar(heat, shrink=0.5, aspect=5)
plt.show()
plt.savefig('u2_a2_e4c')
```

### Solido de revolución acotado por y=1/x, y=0 con x: [1:2] alrededor de y

```python
%matplotlib notebook
box=np.e # set draw area as box^3
fig = plt.figure(figsize=(7,7)) # set fig with its size 10 x 10 inch
ax = fig.add_subplot(111,projection='3d') # creates an additional axis to the standard 2D
# ax.set_xlim(-box,box) # set x-range
# ax.set_ylim(-box,box) # set y-range
# ax.set_zlim(-box,box) # set z-range


# Create the mesh in polar coordinates and compute corresponding Z.
r = np.linspace(1, 2, 50)
p = np.linspace(0,2*np.pi, 50)
R, P = np.meshgrid(r, p)
Z = 1/R
# Express the mesh in the cartesian system.
X, Y = R*np.cos(P), R*np.sin(P)
#
r1 = np.linspace(1, 2, 50)
p1 = np.linspace(0, 2*np.pi, 50)
R1, P1 = np.meshgrid(r1, p1)
X1, Y1 = R1*(np.cos(P1)), R1*(np.sin(P))
Z1= 0*Y1+0*X1


# Plot the surface.
heat=ax.plot_surface(X1, Y1, Z1, cmap=plt.cm.gist_heat, alpha=1)
surf=ax.plot_surface(X, Y, Z, cmap=plt.cm.RdGy, alpha=0.55)

# Tweak the limits and add latex math labels.
#ax.set_zlim(0, 1)
ax.set_xlabel(r"x",fontsize=20) # set x-label
ax.set_ylabel(r"z",fontsize=20) # set y-label
ax.set_zlabel(r"y",fontsize=20) # set z-label

# Add a color bar which maps values to colors.
# fig.colorbar(heat, shrink=0.5, aspect=5)
# fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
plt.savefig('u2_a2_e6a')
```

### Solido de revolución acotado por x+y=3, x=4-(y-1)^2 con x: [1:2] alrededor de x
